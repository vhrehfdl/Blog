# Version1
가장 basic한 형태의 TF-IDF 입니다.

```
from sklearn.feature_extraction.text import TfidfVectorizer

sentence_list = ['오늘은 즐거운 하루입니다', '오늘은 비가 올 것 같은데 내일은 비가 안 왔으면 좋겠네요.', '내일도 비가 올 것 같아요' ]

tfidfv = TfidfVectorizer().fit(sentence_list)
print(tfidfv.transform(sentence_list).toarray())
print(tfidfv.vocabulary_)
```

# Version2
한 글자인 단어들도 Vector에 포함.

```
tfidfv2 = TfidfVectorizer(analyzer='word', token_pattern=r'\w{1,}').fit(sentence_list)
print(tfidfv2.transform(sentence_list).toarray())
print(tfidfv2.vocabulary_)
```

# Version3
위의 TF-IDF에서 ngram_range=(2, 3)를 추가하였습니다.  
단어를 2글자 or 3글자 단위로 묶어 처리하였습니다.

```
tfidfv3 = TfidfVectorizer(analyzer='word', token_pattern=r'\w{1,}', ngram_range=(2, 3)).fit(sentence_list)
print(tfidfv3.transform(sentence_list).toarray())
print(tfidfv3.vocabulary_)
```

# Version4
ngram_range=(2, 3)를 추가하고 analyzer='char'로 변경하였습니다.  
char로 변경하였기 때문에 단어가 아닌 글자가 기준이 됩니다.

```
tfidfv4 = TfidfVectorizer(analyzer='char', token_pattern=r'\w{1,}', ngram_range=(2, 3)).fit(sentence_list)
print(tfidfv4.transform(sentence_list).toarray())
print(tfidfv4.vocabulary_)
```
